```python
import cv2

# Ouvrir la camra
cap = cv2.VideoCapture(0)

# Vrifier si la camra est ouverte correctement
if not cap.isOpened():
    print("Erreur: Impossible d'ouvrir la camra.")
    exit()

# Boucle de capture vido
while True:
    # Lire une image depuis la camra
    ret, frame = cap.read()

    # Afficher l'image
    cv2.imshow('Camera', frame)

    # Attendre 1 milliseconde et vrifier si l'utilisateur appuie sur la touche 'q' pour quitter
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Librer la ressource de la camra et fermer la fentre
cap.release()
cv2.destroyAllWindows()
```
