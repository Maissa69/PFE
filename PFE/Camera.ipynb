```python
import cv2

# Ouvrir la caméra
cap = cv2.VideoCapture(0)

# Vérifier si la caméra est ouverte correctement
if not cap.isOpened():
    print("Erreur: Impossible d'ouvrir la caméra.")
    exit()

# Boucle de capture vidéo
while True:
    # Lire une image depuis la caméra
    ret, frame = cap.read()

    # Afficher l'image
    cv2.imshow('Camera', frame)

    # Attendre 1 milliseconde et vérifier si l'utilisateur appuie sur la touche 'q' pour quitter
    if cv2.waitKey(1) & 0xFF == ord('q'):
        break

# Libérer la ressource de la caméra et fermer la fenêtre
cap.release()
cv2.destroyAllWindows()
```
